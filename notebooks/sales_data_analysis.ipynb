Outlier Detection

```python
df_clean.describe()
```

```python

outlier_summary = {}

for col in df_clean.select_dtypes(include=np.number).columns:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    mask = (df_clean[col] < lower) | (df_clean[col] > upper)

    outlier_summary[col] = mask.sum()

outlier_summary
```

```python

df_clean.select_dtypes(include=np.number).plot(
    kind='box',
    figsize=(10,5)
)

plt.title("Boxplot of Numerical Columns")
plt.show()
```

```python

priority_mapping = {
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Critical': 4,
    'Not Specified': 0
}


df_analysis['priority_numeric'] = df_analysis['priority'].map(priority_mapping)


corr_data = df_analysis[['sales_value', 'priority_numeric']]

correlation_matrix = corr_data.corr()


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap: Priority vs Sales Value")

plt.show()
```
